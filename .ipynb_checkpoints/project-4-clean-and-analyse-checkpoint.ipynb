# Data loading and cleaning

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./seek_lillian.csv', index_col = 0)
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data.salary_amount_searchpage[44]
```

```python
data.salary_amount_searchpage.dropna().str.replace(',','').str.extractall('\$(\d{5,})').unstack()
```

```python
# one job with dodgy numbers
data.salary_amount_searchpage.dropna().str.replace(',','').str.extractall('\$(\d{5,})').unstack().iloc[:, 2:4][0].dropna()
```

```python
# dropped that job
data.drop([16291], inplace=True)
```

```python
data['lower_range'] = data.salary_amount_searchpage.dropna().str.replace(',','').str.extractall('\$(\d{5,})').unstack().iloc[:,0:1]
data['upper_range'] = data.salary_amount_searchpage.dropna().str.replace(',','').str.extractall('\$(\d{5,})').unstack().iloc[:,1:2]
data['salaries_k'] = data.salary_amount_searchpage.dropna().str.replace(',','').str.extractall('-\s?\$(\d\d+)[Kk]').unstack().iloc[:,0:1]
data['salaries_k_real'] = 1000*data.salaries_k.dropna().astype(int)
```

```python
def remove_html_tags(text):
    """Remove html tags from a string"""
    import re
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

def clean_description(raw_desc):
    from bs4 import BeautifulSoup
    soup = BeautifulSoup(raw_desc, features='html.parser')
    cleaned_desc = remove_html_tags(str(soup).replace(',',' ').replace('[','').replace(']','').replace('\n', ' ').replace('.',' '))
    return cleaned_desc
```

```python
cleaned_descriptions = []
counter = 0
for x in data.long_description:
    counter += 1
    print('Almost done? %d' % counter, end='\r')
    try:
        cleaned_descriptions.append(clean_description(x))
    except:
        cleaned_descriptions.append(np.nan)
```

```python
data['cleaned_descriptions'] = cleaned_descriptions
```

```python
data.head(60)
```

```python
range_averages = (data.lower_range.dropna().astype(int) + data.upper_range.dropna().astype(int))/2
```

```python
data['range_averages'] = range_averages.fillna(data.lower_range)
```

```python
data['salary'] = data.salaries_k_real.fillna(data.range_averages)
```

```python
new_data = data.dropna(subset=['salary', 'cleaned_descriptions'])[['job_title', 'company_name','short_description', 'job_location', 'job_category', 'salary', 'cleaned_descriptions']].dropna()
```

```python
new_data.info()
```

```python
import missingno as msno
msno.matrix(new_data)
```

```python
new_data['lowmidhigh_salaries'] = pd.qcut(new_data.salary.astype(int), 3, ['low', 'mid', 'high'])
```

```python
new_data['lowhigh_salaries'] = pd.qcut(new_data.salary.astype(int), 2, ['low', 'high'])
```

# Job title modelling

## Regression

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression, LassoCV, RidgeCV
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
import eli5
import wordcloud
import matplotlib.pyplot as plt
import shap
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
X = new_data.job_title
y = new_data.salary.astype(int)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5)
```

```python
model = make_pipeline(CountVectorizer(ngram_range=(1,2)), LinearRegression())

model.fit(X_train, y_train)
model.score(X_test, y_test), model.score(X_train, y_train)
np.mean(cross_val_score(model, X_test, y_test))
```

Basic Linear regression model had a terrible score, so i'm going to disregard it and try ridge and lasso.

```python
model1 = make_pipeline(CountVectorizer(ngram_range=(1,2), binary=True), RidgeCV(cv=5))

model1.fit(X_train, y_train)
print(model1.score(X_test, y_test), cross_val_score(model1, X, y))
```

```python
eli5.explain_weights(model1, top=40)
```

```python
model2 = make_pipeline(CountVectorizer(ngram_range=(1,2), binary=True), LassoCV(n_jobs=-1, verbose=1))

model2.fit(X, y)
print(cross_val_score(model2, X, y))
```

RidgeCV had a better score than Lasso (which also took forever to run), so i have visualised the ridgecv model.

It appears that with overwhelming support, the features that best predict salary in data jobs are seniority precursers to the job title such as 'senior', 'lead', 'president', 'director', and on the lower range 'assistant'. 'BI' jobs also tend to be high salary roles, and knowledge in saas increases expected salary too.

## Classification with low and high salary groups only 

Just for fun, i also tried classification with only two groups, low and high salary.

```python
X = new_data.job_title
y = new_data.lowhigh_salaries
corp_train, corp_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

```python
model7 = make_pipeline(CountVectorizer(ngram_range=(1,2), binary=True), LogisticRegression())

model7.fit(X, y)

print(np.mean(cross_val_score(model7, X, y)), cross_val_score(model7, X, y))
```

```python
shap.summary_plot(shap_values, X_test_array, feature_names=vectorizer.get_feature_names(), plot_type='violin', max_display=30)
```

```python
eli5.explain_weights(model7, top=40)
```

```python
model8 = make_pipeline(CountVectorizer(ngram_range=(1,2), binary=True), RandomForestClassifier())

model8.fit(X, y)

print(np.mean(cross_val_score(model8, X, y)), cross_val_score(model8, X, y))
```

```python
y.value_counts(normalize=True)
```

Not a lot of new information to be derived here, but the score of the logistic regression model with two classes performs a lot better than the same with three. Also, Randomforestclassifier has consistently performed worse than logistic regression throughout all the previous analysis.

# Question 2 Job category

I've chosen to investigate the differences in job demand and differences in salary between data jobs in Sydney and Melbourne. 

Differences in job demand will be evaluated using the features in the job titles and long descriptions of roles (which will hopefully pick up on any differences in skills required, and/or seniority). Job demand will also be evaluated using the 'job category' column, which will give some insight into the types of industries that are currently hiring in each city. 

Differences in salary will be evaluated using a t-test between Sydney and Melbourne salaries using the salary column.

```python
syd_mel = new_data[(new_data.job_location == 'Sydney')| (new_data.job_location == 'Melbourne')]
len(new_data[new_data.job_location == 'Sydney']), len(new_data[new_data.job_location == 'Melbourne'])
```

## Job titles

```python
X = syd_mel.job_title
y = syd_mel.job_location
```

```python
model11 = make_pipeline(CountVectorizer(ngram_range=(1,2), binary=True, stop_words='english'), LogisticRegression(max_iter=1000))

model11.fit(X, y)

print(np.mean(cross_val_score(model11, X, y)), cross_val_score(model11, X, y))
```

```python
eli5.explain_weights(model11, top=40)
```

```python
model12 = make_pipeline(CountVectorizer(ngram_range=(1,2), binary=True, stop_words='english'), RandomForestClassifier())

model12.fit(X, y)

print(np.mean(cross_val_score(model12, X, y)), cross_val_score(model12, X, y))
```

From the job titles only, it appears that there is a difference between Sydney and Melbourne. There is a lack of the use of "senior" and "junior" precurser terms in the job titles advertised in Melbourne, instead there is usage of "head" and "coordinator". Marketing work seems to be important for both cities and sought after. It's possible that Melbourne job titles are more descriptive and specific to actual skills required.

Base RandomForestClassifier performed worse on this data than base LogisticRegression, same as before although the rfc appeared to have a better fit this time (less variance in cross val score values).
